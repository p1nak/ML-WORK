```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('job_salary_prediction_dataset.csv')
```

```python
df.columns
```

```python
def plotting(var,num):
    plt.subplot(2,2,num)
    sns.histplot(df[var], kde=True)
plotting('experience_years',1)
plotting('skills_count',2) 
plotting('certifications',3) 
plotting('salary',4) 
plt.tight_layout()
```

```python
numerical_value = ['experience_years','skills_count','certifications','salary']
for col in numerical_value:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
```

```python
df['job_title'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df['job_title'], y=df['salary'])

plt.xticks(rotation=45, ha='right')  # rotate labels
plt.tight_layout()
plt.show()
```
