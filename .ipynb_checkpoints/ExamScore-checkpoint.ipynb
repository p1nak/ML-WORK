```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

```python
df = df.astype(int)
```

```python
numerical_value = [
    "Hours_Studied",
    "Attendance",
    "Sleep_Hours",
    "Previous_Scores",
    "Tutoring_Sessions",
    "Physical_Activity",
    "Exam_Score"
]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numerical_value] = scaler.fit_transform(df[numerical_value])
df.head()
```

```python
df.columns
```

```python
from scipy.stats import pearsonr

selected_features = [
'Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores',
       'Tutoring_Sessions', 'Physical_Activity', 
       'Parental_Involvement_Low', 'Parental_Involvement_Medium',
       'Access_to_Resources_Low', 'Access_to_Resources_Medium',
       'Extracurricular_Activities_Yes', 'Motivation_Level_Low',
       'Motivation_Level_Medium', 'Internet_Access_Yes', 'Family_Income_Low',
       'Family_Income_Medium', 'Teacher_Quality_Low', 'Teacher_Quality_Medium',
       'School_Type_Public', 'Peer_Influence_Neutral',
       'Peer_Influence_Positive', 'Learning_Disabilities_Yes',
       'Parental_Education_Level_High School',
       'Parental_Education_Level_Postgraduate', 'Distance_from_Home_Moderate',
       'Distance_from_Home_Near', 'Gender_Male'

]

correlations = {
    feature: pearsonr(df[feature], df['Exam_Score'])[0]
    for feature in selected_features
}
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson Correlation'])

correlation_df.sort_values(by='Pearson Correlation', ascending=False)
```

```python
cat_features = [
    'Parental_Involvement_Low',
    'Parental_Involvement_Medium',
    'Access_to_Resources_Low',
    'Access_to_Resources_Medium',
    'Extracurricular_Activities_Yes',
    'Motivation_Level_Low',
    'Motivation_Level_Medium',
    'Internet_Access_Yes',
    'Family_Income_Low',
    'Family_Income_Medium',
    'Teacher_Quality_Low',
    'Teacher_Quality_Medium',
    'School_Type_Public',
    'Peer_Influence_Neutral',
    'Peer_Influence_Positive',
    'Learning_Disabilities_Yes',
    'Parental_Education_Level_High School',
    'Parental_Education_Level_Postgraduate',
    'Distance_from_Home_Moderate',
    'Distance_from_Home_Near',
    'Gender_Male'
]
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05
chi2_results = {}

target = df['Exam_Score']

for col in cat_features:
    contingency = pd.crosstab(df[col], target)
    
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')

chi2_df
```
