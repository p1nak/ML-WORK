```python
from sklearn.preprocessing import StandardScaler
numerical_cols = ['cgpa', 'backlogs','internship_count'	,'aptitude_score','communication_score','internship_quality_score',	'salary']
scaler = StandardScaler()
df_encode[numerical_cols] = scaler.fit_transform(df_encode[numerical_cols])
df_encode.head()
```

```python
df_encode.columns
```

```python
from scipy.stats import pearsonr

selected_features = [
    'cgpa', 'backlogs', 'internship_count', 'aptitude_score',
       'communication_score', 'internship_quality_score', 'salary',
       'college_tier_Tier 2', 'college_tier_Tier 3', 'country_Germany',
       'country_India', 'country_UK', 'country_USA',
       'university_ranking_band_300+', 'university_ranking_band_Top 100',
       'specialization_Cloud', 'specialization_Core CS',
       'specialization_Cybersecurity', 'specialization_Data Science',
       'industry_Finance', 'industry_Healthcare', 'industry_Manufacturing',
       'industry_Other', 'industry_Tech'

]

correlations = {
    feature: pearsonr(df_encode[feature], df_encode['placement_status_Placed'])[0]
    for feature in selected_features
}
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson Correlation'])
correlation_df.sort_values(by='Pearson Correlation', ascending=False)
```

```python
cat_features = [
       'college_tier_Tier 2', 'college_tier_Tier 3', 'country_Germany',
       'country_India', 'country_UK', 'country_USA',
       'university_ranking_band_300+', 'university_ranking_band_Top 100',
       'specialization_Cloud', 'specialization_Core CS',
       'specialization_Cybersecurity', 'specialization_Data Science',
       'industry_Finance', 'industry_Healthcare', 'industry_Manufacturing',
       'industry_Other', 'industry_Tech', 'placement_status_Placed']
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05
chi2_results = {}

target = df_encode['placement_status_Placed']

for col in cat_features:
    contingency = pd.crosstab(df_encode[col], target)
    
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')

chi2_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_encode.drop('salary',axis=1)
y = df_encode['salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred) 
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1- r2) * (n-1) / (n - p - 1))
adjusted_r2
```
